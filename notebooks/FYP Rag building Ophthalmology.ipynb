# PDF to Markdown

```python
import pymupdf.layout
import pymupdf4llm
from pathlib import Path
```

```python
# Testing with this

import pymupdf
import pymupdf4llm
from pathlib import Path

input_file = "Ophthalmology.pdf"
output_file = "ophthalmology_knowledge_base.md"

doc = pymupdf.open(input_file)
toc = doc.get_toc() # This is the "Gold Standard" for your hierarchy

# 1. Create a mapping of Page Number -> Heading Level
# This helps us know exactly which '##' should actually be '###'
toc_map = {}
for level, title, page in toc:
    if page not in toc_map:
        toc_map[page] = []
    toc_map[page].append({"level": level, "title": title})

def process_with_hierarchy():
    Path(output_file).write_text("", encoding="utf-8")

    # We use pymupdf4llm's page-by-page processing for better control
    for page_num in range(len(doc)):
        if page_num % 100 == 0:
          print(f"Processing page {page_num+1} to {page_num+100}...")

        # Get the markdown for this specific page
        page_md = pymupdf4llm.to_markdown(doc, pages=[page_num])

        # If this page has a TOC entry, we can force-adjust the markers
        if (page_num + 1) in toc_map:
            for entry in toc_map[page_num + 1]:
                # Map PDF Level 1 -> #, Level 2 -> ##, Level 3 -> ###
                # This fixes the "everything is a ##" issue generically
                new_marker = "#" * entry['level']
                # Search for the title and ensure it has the correct level
                page_md = page_md.replace(f"## {entry['title']}", f"{new_marker} {entry['title']}")

        with open(output_file, "a", encoding="utf-8") as f:
            f.write(page_md + "\n\n")

process_with_hierarchy()
doc.close()
```

# Cleaning

```python
import re
```

```python
def clean_medical_markdown(text):
    # 1. Remove the specific book title heading (# KANSKI’S Clinical Ophthalmology)
    # This matches the # followed by the title, accounting for potential extra spaces
    text = re.sub(r"^#\s*KANSKI[’']S\s*Clinical\s*Ophthalmology.*?\n", "", text, flags=re.MULTILINE | re.IGNORECASE)

    # 2. Remove the "picture intentionally omitted" blocks
    text = re.sub(r"\*\*==> picture \[.*?\] intentionally omitted <==\*\*", "", text)

    # 3. Remove the "Start of picture text" blocks including markers and content inside
    text = re.sub(r"\*\*----- Start of picture text -----\*\*.*?\*\*----- End of picture text -----\*\*", "", text, flags=re.DOTALL)

    # 4. Remove the passfans URL and its variations
    #text = re.sub(r"(##\s+)?www\.passfans\.com/forum", "", text)

    # 5. Clean up excess newlines left behind (3 or more becomes 2)
    text = re.sub(r'\n{3,}', '\n\n', text)

    return text.strip()
```

```python
md_path="ophthalmology_knowledge_base.md"
with open(md_path, "r", encoding="utf-8") as f:
    md_text = f.read()
```

```python
cleaned_md = clean_medical_markdown(md_text)
```

```python
def restructure_medical_markdown(text):
    # 1. Promote current ## to #
    # This makes major topics the top-level parent.
    text = re.sub(r'^##\s+', '# ', text, flags=re.MULTILINE)

    # 2. Promote current ### to ##
    # This makes specific diseases/conditions the second-level parent.
    text = re.sub(r'^###\s+', '## ', text, flags=re.MULTILINE)

    # 3. Convert '- **......**' to ###
    # This turns your bolded list items into proper third-level headers.
    # It handles both '- **Topic**' and '- **Topic.**' (with optional punctuation).
    text = re.sub(r'^\s*-\s*\*\*(.*?)\*\*[\.\s]*', r'### \1\n', text, flags=re.MULTILINE)

    # 4. Final Cleanup: Standardize spacing and remove noise
    text = re.sub(r'\n{3,}', '\n\n', text)

    return text.strip()
```

```python
cleaned_md = restructure_medical_markdown(cleaned_md)
```

# Chunking

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter, RecursiveCharacterTextSplitter

# 1. Define the headers to track (based on your medical md structure)
# This attaches the section titles to the chunk's metadata automatically.
headers_to_split_on = [
    ("#", "Header_1"),
    ("##", "Header_2"),
    ("###", "Header_3"),
]

# 2. Initialize the Header Splitter
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on=headers_to_split_on)

# md_text is your cleaned variable from the previous step
md_header_splits = markdown_splitter.split_text(cleaned_md)

# 3. Further split these sections into manageable chunks
# A 1000-character chunk with 10-15% overlap is the "sweet spot" for medical data.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=150,
    separators=["\n\n", "\n", ". ", " ", ""] # Priority: Paras > Lines > Sentences > Words
)

# 4. Execute the final recursive split
final_chunks = text_splitter.split_documents(md_header_splits)

# --- Verification ---
print(f"Total Chunks Created: {len(final_chunks)}")
print("\n--- Sample Chunk Metadata (Structure Context) ---")
print(final_chunks[10].metadata)  # Shows which headers this chunk belongs to
print("\n--- Sample Chunk Content ---")
print(final_chunks[10].page_content[:200] + "...")
```

## Saving chunks

```python
import json

def save_chunks_to_jsonl(chunks, filename):
    with open(filename, 'w', encoding='utf-8') as f:
        for chunk in chunks:
            # Convert LangChain Document to a serializable dict
            line = {
                "page_content": chunk.page_content,
                "metadata": chunk.metadata
            }
            f.write(json.dumps(line) + '\n')
    print(f"Successfully saved {len(chunks)} chunks to {filename}")

# Usage
save_chunks_to_jsonl(final_chunks, "ophthalmology_medical_chunks.jsonl")
```

```python
import json
from langchain_core.documents import Document

def load_chunks_from_jsonl(filename):
    loaded_chunks = []
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            # This recreates the LangChain Document object exactly as it was
            doc = Document(
                page_content=data["page_content"],
                metadata=data["metadata"]
            )
            loaded_chunks.append(doc)
    print(f"Successfully loaded {len(loaded_chunks)} chunks from {filename}")
    return loaded_chunks

# Usage: Load them back into the variable 'final_chunks'
final_chunks = load_chunks_from_jsonl("ophthalmology_medical_chunks.jsonl")

# --- Verification (Check if it looks the same as before) ---
print(f"Total Chunks Loaded: {len(final_chunks)}")
print(f"Sample Metadata: {final_chunks[10].metadata}")
```

# Embeddings

```python
from google.colab import userdata
```

```python
import os
import time
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma

# 1. Setup API Key
os.environ["GOOGLE_API_KEY"] = userdata.get('GOOGLE_API_KEY')

# 2. Initialize Gemini Embeddings
embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")

# 3. Create/Initialize the Vector Store (empty at first)
# We initialize it once, then add documents in a loop
vector_db = Chroma(
    persist_directory="./ophthalmology_doctorhive_gemini_db",
    embedding_function=embeddings
)

# 4. Batch Processing Configuration
batch_size = 100  # Number of chunks per API call
delay_seconds = 2 # Wait time to avoid rate limits

print(f"Starting batch processing of {len(final_chunks)} chunks...")

for i in range(0, len(final_chunks), batch_size):
    # Grab a slice of chunks
    batch = final_chunks[i : i + batch_size]

    # Add this specific batch to the database
    vector_db.add_documents(batch)

    # Progress Update
    print(f"Processed chunks {i} to {min(i + batch_size, len(final_chunks))}...")

    # The essential delay
    time.sleep(delay_seconds)

print("\nAll chunks successfully embedded and saved!")
```

```python
embeddings = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004")
vector_db = Chroma(
    persist_directory="./ophthalmology_doctorhive_gemini_db",
    embedding_function=embeddings
)

# 2. Define a medical query
query = "What are the common symptoms of cataracts"

# 3. Perform Similarity Search
# k=3 means "give me the top 3 most relevant chunks"
results = vector_db.similarity_search(query, k=3)

# 4. Inspect the results
```
